```python
import pandas as pd
```

```python
#Generate an ID mapping dictionary for gene symbol and NCBI gene ID
ID_map = pd.read_csv("/Users/guangrong/Documents/GitHub_project/fastqpi_BigGIM/KGs/ID_convert/Table_ID_convert_9.19.2021.tsv", sep = '\t')
dic_ID_symbol2ncbi ={}
import math
ID_map.astype(str)
for i in range(0,ID_map.shape[0]):
    symbol = ID_map['Approved_symbol'][i]
    
    ncbi_id = ID_map['NCBI_Gene_ID'][i]
    ensemble_id = ID_map['Ensembl_gene_ID'][i]
    if math.isnan(ncbi_id) == False :
        ncbi_id = int(ncbi_id)
        if symbol not in dic_ID_symbol2ncbi:
            dic_ID_symbol2ncbi[symbol] = ncbi_id
        else:
            print(symbol)
for i in range(0,ID_map.shape[0]):
    alias_list = str(ID_map["Alias_symbols"][i]).split(',')
    ncbi_id = ID_map['NCBI_Gene_ID'][i]
    ensemble_id = ID_map['Ensembl_gene_ID'][i]
    if len(alias_list) > 0:
        for alias in alias_list:
            if alias not in dic_ID_symbol2ncbi:
                dic_ID_symbol2ncbi[alias] = ncbi_id
dic_ID_symbol2ncbi['MEN1'] = 4221
dic_ID_symbol2ncbi['TSC1']= 7248
dic_ID_symbol2ncbi['WHSC1'] = 7468
dic_ID_symbol2ncbi['MET'] = 4233
dic_ID_symbol2ncbi['HIST1H1E'] = 3008
dic_ID_symbol2ncbi['FLNA'] = 2316
dic_ID_symbol2ncbi["FAM46D"] = 169966
dic_ID_symbol2ncbi["HIST1H1C"] = 3006
```

```python
subject_name = ["FANCA", "FANCC", "BRCA2", "FANCD2",
                "FANCE", "FANCF", "FANCG", "FANCI", "BRIP1",
                "FANCL", "FANCM", "PALB2", "RAD51C", "SLX4", 
                "ERCC4", "BRCA1", "UBE2T", "XRCC2", 
                "MAD2L2", "RFWD3", "RAD51"]
subject_id_list = []
for name in subject_name:
    if name in dic_ID_symbol2ncbi:
        subject_id_list.append(dic_ID_symbol2ncbi[name])
    else:
        print(name)

subject_id_prefixes = ['NCBIGene'] * len(subject_name)
subject_category = ["Gene"] * len(subject_name)

object_name = ["Fanconi anemia"] * len(subject_name)
object_id = ["MONDO:0019391"] * len(subject_name)
object_id_prefix = ["MONDO"] * len(subject_name)
object_category = ["Disease"] * len(subject_name)

predicate = ["gene_associated_with_condition"] * len(subject_name)
subject_aspect_qualifier = ['Genetic variants'] * len(subject_name)
knowledge_source = ["PMID: 35854323"] * len(subject_name)
```

```python
FA_mut_grap = pd.DataFrame()
FA_mut_grap['subject_name'] = subject_name
FA_mut_grap['subject_id_prefix'] =subject_id_prefixes
FA_mut_grap['subject_id'] = subject_id_list
FA_mut_grap['subject_category'] =subject_category

FA_mut_grap['object_name'] = object_name
FA_mut_grap['object_id'] = object_id
FA_mut_grap['object_id_prefix'] = object_id_prefix
FA_mut_grap['object_category'] = object_category

FA_mut_grap['predicate'] = predicate
FA_mut_grap['subject_aspect_qualifier'] = subject_aspect_qualifier
FA_mut_grap['knowledge_source'] =knowledge_source
```
