```python
#Generate an ID mapping dictionary for gene symbol and NCBI gene ID
ID_map = pd.read_csv("/Users/guangrong/Documents/GitHub_project/fastqpi_BigGIM/KGs/ID_convert/Table_ID_convert_9.19.2021.tsv", sep = '\t')
dic_ID_symbol2ncbi ={}
dic_ID_ensemble2ncbi = {}
dic_ID_ncbi2symbol = {}
import math
ID_map.astype(str)

for i in range(0,ID_map.shape[0]):
    
    symbol = ID_map['Approved_symbol'][i]
    ncbi_id = ID_map['NCBI_Gene_ID'][i]
    if math.isnan(ncbi_id) == False :
        ncbi_id = int(ncbi_id)
        if ncbi_id not in dic_ID_ncbi2symbol:
            dic_ID_ncbi2symbol[ncbi_id] = symbol
        else:
            print(ncbi_id)
    
for i in range(0,ID_map.shape[0]):
    symbol = ID_map['Approved_symbol'][i]
    
    ncbi_id = ID_map['NCBI_Gene_ID'][i]
    ensemble_id = ID_map['Ensembl_gene_ID'][i]
    if math.isnan(ncbi_id) == False :
        ncbi_id = int(ncbi_id)
        if symbol not in dic_ID_symbol2ncbi:
            dic_ID_symbol2ncbi[symbol] = ncbi_id
            #dic_ID_ncbi2symbol[ncbi_id] = symbol
        else:
            print(symbol)
        
        if ensemble_id not in dic_ID_ensemble2ncbi:
            dic_ID_ensemble2ncbi[ensemble_id] = ncbi_id
        else:
            print(ensemble_id)
           # print(ncbi_id)
            
for i in range(0,ID_map.shape[0]):
    alias_list = str(ID_map["Alias_symbols"][i]).split(',')
    ncbi_id = ID_map['NCBI_Gene_ID'][i]
    ensemble_id = ID_map['Ensembl_gene_ID'][i]
    if len(alias_list) > 0:
        for alias in alias_list:
            if alias not in dic_ID_symbol2ncbi:
                dic_ID_symbol2ncbi[alias] = ncbi_id
                dic_ID_ncbi2symbol[ncbi_id] = alias
                
dic_ID_symbol2ncbi['MEN1'] = 4221
dic_ID_symbol2ncbi['TSC1']= 7248
dic_ID_symbol2ncbi['WHSC1'] = 7468
dic_ID_symbol2ncbi['MET'] = 4233
dic_ID_symbol2ncbi['HIST1H1E'] = 3008
dic_ID_symbol2ncbi['FLNA'] = 2316
dic_ID_symbol2ncbi["FAM46D"] = 169966
dic_ID_symbol2ncbi["HIST1H1C"] = 3006
```

```python
HuRI_data = pd.read_csv(input_dir1 + "HI-union.tsv", sep ='\t',  header = None)
HuRI_data.columns = ['G1','G2']
missed_id = set()
for i in HuRI_data['G1']:
    if i not in dic_ID_ensemble2ncbi:
        missed_id.add(i)
for i in HuRI_data['G2']:
    if i not in dic_ID_ensemble2ncbi:
        missed_id.add(i)
len(missed_id)
```

```python
Result = pd.DataFrame()
subject_id_list = []
object_id_list = []

for i in range(0,HuRI_data.shape[0]):
    if HuRI_data.iloc[i,0] not in missed_id and HuRI_data.iloc[i,1] not in missed_id:
        subject_id_list.append(dic_ID_ensemble2ncbi[HuRI_data.iloc[i,0]])
        object_id_list.append(dic_ID_ensemble2ncbi[HuRI_data.iloc[i,1]])

subject_name_list = []
object_name_list = []
for i in subject_id_list:
    subject_name_list.append(dic_ID_ncbi2symbol[i])
for j in object_id_list:
    object_name_list.append(dic_ID_ncbi2symbol[j])

Result = pd.DataFrame()
Result['subject_id'] = subject_id_list
Result['object_id'] = object_id_list
Result['subject_id_prefix'] = ['NCBIGene'] * len(subject_id_list)
Result['object_id_prefix'] = ['NCBIGene'] * len(object_id_list)
Result['subject_name'] = subject_name_list
Result['object_name'] = object_name_list
Result['predicate'] = ["physically_interacts_with"] * len(subject_id_list)
Result['knowledge_source'] = ['HuRI'] * len(subject_id_list)
Result['publications'] = ['PMID: 32296183'] * len(subject_id_list)
Result['subject_category'] = ['Protein'] * len(subject_id_list)
Result['object_category'] = ['Protein'] * len(subject_id_list)
Result.to_csv("../Input_Table/HuRI_formated.csv", index = None)
```
